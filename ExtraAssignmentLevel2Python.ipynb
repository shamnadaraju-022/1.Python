```python
# print Odd Numbers in the list - (20,10,16,19,25,1,276,188) 
19 is odd
25 is odd
1 is odd
```

```python
numbers = (20, 10, 16, 19, 25, 1, 276, 188)
for num in numbers:
        if num % 2 != 0:
            print(num, "is odd")
```

```python
# print Even numbers in the list (20,10,16,19,25,1,276,188)
20 is odd
10 is odd
16 is odd
276 is odd
188 is odd
```

```python
numbers = (20,10,16,19,25,1,276,188)
for num in numbers:
    if num % 2 == 0: 
        print(num, "is even")
```
